```python

# === Import Libraries ===
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python

# === Load the Pickle File ===
file_path = r"C:\Users\malak\OneDrive - Morgan State University\Team 8 Group Project Folder\WESAD\data_stress.pkl"

file = pd.read_pickle(file_path)

# verify data is loaded
print("===== Dataset Overview =====")
print("Shape:", file.shape)
print("\n===== Info =====")
print(file.info())
```

```python

# === Display Column Names ===
print("===== Columns =====")
for i, col in enumerate(file.columns, 1):
    print(f"{i}. {col}")
```

```python

# === Label Distribution ===
plt.figure(figsize=(6,4))
sns.countplot(x="Stress Levels", data=file)
plt.title("Label Distribution")
plt.show()

print("\nLabel Counts:")
print(file["Stress Levels"].value_counts())
```

```python

# === Feature Distributions ===
numeric_features = file.select_dtypes(include='number').columns.tolist()

# Remove label column if numeric
if "Stress Levels" in numeric_features:
    numeric_features.remove("Stress Levels")

for feature in numeric_features:
    plt.figure(figsize=(6,4))
    sns.histplot(file[feature], bins=30, kde=True, color="blue")
    plt.title(f"{feature} Distribution")
    plt.show()
```

```python

# === Correlation Heatmap ===
plt.figure(figsize=(10,6))
sns.heatmap(file.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()

# Correlates to stress when the # is closest to 1
# opposite Correlations when close to -1
# no corelation when clsoe to 0
```
